```python
import lxml
import pandas as pd 
import os 
from glob import glob 
```

```python
import xml.etree.ElementTree as ET
```

```python
def get_coordinates_for_xml(xml_file, dataframe):
    sample = 'Marmot_data/10.1.1.1.2006_3.xml'
    tree = ET.parse(sample)
    root = tree.getroot()
    boxes = root.findall("object/bndbox")
    
    file_root_name = "".join(xml_file.split('/')[-1])[:-4]
    
    for box in boxes:
        x_min = box.find('xmin').text
        y_min = box.find('ymin').text
        x_max = box.find('xmax').text
        y_max = box.find('ymax').text
    
#         print(x_min, y_min, x_max, y_max)
        row = pd.Series([file_root_name, x_min, y_min, x_max, y_max])
        dataframe = dataframe.append(row, ignore_index=True)
        
    return dataframe
```

```python
xml_files = glob("Marmot_data/*.xml")
print(len(xml_files))
```

```python
df = pd.DataFrame()
for file in xml_files:
    df = get_coordinates_for_xml(file, df)
```
