```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sample.jpg')
img = img[:,:,::-1]
```

```python
plt.imshow(img)
```

```python
size = 40
replicated = cv2.copyMakeBorder(img, size, size, size, size, cv2.BORDER_REPLICATE)
reflected = cv2.copyMakeBorder(img, size, size, size, size, cv2.BORDER_REFLECT)
reflected101 = cv2.copyMakeBorder(img, size, size, size, size, cv2.BORDER_REFLECT_101)
wrapped = cv2.copyMakeBorder(img, size, size, size, size, cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img, size, size, size, size, cv2.BORDER_CONSTANT, value=[255, 255, 255]) 
```

```python
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(235)
ax6 = fig.add_subplot(236)

ax1.imshow(img, cmap='gray')
ax1.title.set_text('Original Image')
ax2.imshow(replicated, cmap='gray')
ax2.title.set_text('Replicated Border')
ax3.imshow(reflected, cmap='gray')
ax3.title.set_text('Reflected Border')
ax4.imshow(reflected101, cmap='gray')
ax4.title.set_text('Reflected 101 Border')
ax5.imshow(wrapped, cmap='gray')
ax5.title.set_text('Wrapped Border')
ax6.imshow(constant, cmap='gray')
ax6.title.set_text('Constant Border (White)')
plt.show()
```
